```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
credits.head(1)
```

```python
# firstly we will merge these two dataframes into one
movies=movies.merge(credits,on='title')
```

```python
movies.shape
movies.info()
# movies.head(1)

```

```python
# now let us remove the columns which are not in use
movies.info()
```

```python
#feature selection
'''
genre
id
keywords
title
overview
cast
crew
'''
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.shape
```

```python
movies = movies.drop_duplicates()
movies.duplicated().sum()
movies.shape
```

# formatting ther genres column

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L     
```

```python
'''dont run this cell as already converted genre column will throw an error that it is already converted in a list
and ast dont take literal only takes list,only run it one time'''
# movies['genres']=movies['genres'].apply(convert) 
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L  
```

```python
# same for cast and crew but with different function and concvert function
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast'][0:11]
# movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L    
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
# movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
# movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" " .join(x))
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
#Creating a main function now to recommend
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
